```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= pd.read_csv("C:/Rohan Machine Learning/Diabetes-Dataset.csv")
```

```python
data
```

```python
data.isnull().sum()
```

## Spliting data set 

```python
X=data.drop(['Outcome'], axis=1)
print(X)
```

```python
Y = data['Outcome']
print(Y)
```

## X_Y train and test 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
print(X_train.value_counts().sum())
print(X.value_counts().sum())
print("X_test =",X_test.value_counts().sum(),"count")
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#tree_m= DecisionTreeClassifier(max_depth =4)

# Create Decision Tree classifer object
tree_m = DecisionTreeClassifier()
```

```python
tree_m.fit(X_train,Y_train) 
```

```python
from sklearn import tree
plt.figure(figsize=(20,20))
tree.plot_tree(tree_m,filled = True)
           
```

# Prediction 

```python
Y_pred = tree_m.predict(X_test)
```

```python
Y_pred
```

## Matrics

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(Y_pred,Y_test)
print(score*100)
```
